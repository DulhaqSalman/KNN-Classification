```python
Glucose_mean=data.loc[:,"Glucose"].mean()
BloodPressure_mean=data.loc[:,"BloodPressure"].mean()
```

```python
SkinThickness_mdeian=data.loc[:,"SkinThickness"].median()
Insulin_median=data.loc[:,"Insulin"].median()
BMI_median=data.loc[:,"BMI"].median
```

```python
data.loc[:,"Glucose"]=data.loc[:,"Glucose"].fillna(Glucose_mean)
data.loc[:,"BloodPressure"]=data.loc[:,"BloodPressure"].fillna(BloodPressure_mean)
data.loc[:,"SkinThickness"]=data.loc[:,"SkinThickness"].fillna(SkinThickness_mdeian)
data.loc[:,"Insulin"]=data.loc[:,"Insulin"].fillna(Insulin_median)
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler
n=MinMaxScaler()
x_t=n.fit(x_train)
x_t
```

```python
x_train=x_t.transform(x_train)
x_train
```

```python
x_test=x_t.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test,y_pred)
```

```python
knn.score(x_test,y_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(classification_report(y_test,y_pred))
```
